```python
import ipywidgets as widgets
```

```python
c = widgets.Dropdown(
    options=data['p_year'].unique(),
    value=2020,
    description='Select Year',
    disabled=False,
)
d = widgets.Dropdown(
    options=['view_count', 'likes' , 'dislikes'],
    value= 'view_count',
    description='Select',
    disabled=False,
)
t = widgets.Dropdown(
    options=['Barplot', 'Boxplot'],
    value= 'Barplot',
    description='Select Graph',
    disabled=False,
)

def graph(a , z , p):
  if p == 'Barplot':
    sns.barplot(y = 'genre', x = z, data = data[data['p_year'] == a], estimator = sum )
    plt.title(a)
    plt.show()
  if p == 'Boxplot':
    sns.boxplot(x = 'view_count' , y='genre', data =data[data['p_year'] == a])
    plt.title(a)
    plt.show()
```

```python
widgets.interact(graph, a = c, z = d, p = t)               # Created user friendly Drop Dwon menu to select required year and user visualise data based on genre
```
